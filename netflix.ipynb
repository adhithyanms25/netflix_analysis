```python
# 1.which TV show has the most rating
tv_show=df_draft[df_draft["type"]=="TV Show"]
tv_show.sort_values(by="viewer_rating",ascending=False).head(1)
```

```python

# graph of tv show which have more rating
top_tv_shows = tv_show.sort_values(by="viewer_rating", ascending=False).head(10)
x = top_tv_shows['title']
y = top_tv_shows['viewer_rating']
plt.plot(x, y,"d:", color='red')
plt.xlabel('Viewer Rating')
plt.ylabel('TV Show Title')
plt.title('Top 5 TV Shows by Viewer Rating')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 2.How many movies starts with the letter 's'
movies=df_draft[df_draft["type"]=="Movie"]
m_c=movies["title"].str.lower().str.startswith('s',na=False).sum()
print("number of movies start with letter 's': ",m_c)
```

```python
# 3.country which have the most contribution
df_gp_country=df_draft.groupby("country")["title"].count()
country_most_contribution=df_gp_country.idxmax()
most_contribution=df_gp_country.max()
print("country which have the least contribution: ",country_most_contribution)
print("most contribution: ",most_contribution)
```

```python
# country which have the most contribution
x=df_gp_country.sort_values(ascending=False).head(10).index
y=df_gp_country.sort_values(ascending=False).head(10)
plt.plot(x,y,"D-k")
plt.xlabel("country")
plt.ylabel("number of contribution")
plt.title("country which have the most contribution")
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# 4.which years has the most and least movie released
movies=df_draft[df_draft["type"]=="Movie"]
movies_gp_year_most=movies.groupby("release_year")["title"].count().head(1)
movies_gp_year_least=movies.groupby("release_year")["title"].count().tail(1)
print("most movies released in year: ",movies_gp_year_most)
print("least movies released in year: ",movies_gp_year_least)
```

```python
# 8.which TV show and Movie has the most duration
# Filter for movies and TV shows
movies= df_draft[df_draft['type'] == 'Movie'].copy()
tv_shows=df_draft[df_draft['type'] == 'TV Show'].copy()

# Function to extract duration for movies (in minutes)
def get_movie_duration(duration_str):
    if isinstance(duration_str, str) and 'min' in duration_str:
        return int(duration_str.replace(' min', ''))
    return 0

# Function to extract duration for TV shows (in seasons)
def get_tv_show_duration(duration_str):
    if isinstance(duration_str, str):
        if 'Seasons' in duration_str:
            return int(duration_str.replace(' Seasons', ''))
        elif 'Season' in duration_str:
            return int(duration_str.replace(' Season', ''))
    return 0

# Apply the duration extraction
movies['duration_numeric'] = movies['duration'].apply(get_movie_duration)
tv_shows['duration_numeric'] = tv_shows['duration'].apply(get_tv_show_duration)

# Find the movie with the most duration
most_duration_movie = movies.sort_values(by='duration_numeric', ascending=False).iloc[0]
print(f"Movie with the most duration: '{most_duration_movie['title']}' with {most_duration_movie['duration_numeric']} minutes.")

# Find the TV show with the most duration
most_duration_tv_show = tv_shows.sort_values(by='duration_numeric', ascending=False).iloc[0]
print(f"TV Show with the most duration: '{most_duration_tv_show['title']}' with {most_duration_tv_show['duration_numeric']} seasons.")
```

```python
# 9.which genre has the most avg rating and wich country contributed for this genre
df_gp_genre=df_draft.groupby("listed_in")["viewer_rating"].mean()
dgg=df_gp_genre.sort_values(ascending=False).idxmax()
music_df_gp_country=df_draft[df_draft["listed_in"]=="Music & Musicals, Stand-Up Comedy"]
mdgc=music_df_gp_country.value_counts("country").idxmax()
print("genre which has the most avg rating: ",dgg)
print("country which contributed for this genre: ",mdgc)
```

```python
# 10.most rated director in india/other country
india_df = df_draft[df_draft["country"] == "India"]

director_counts = india_df['director'].value_counts()

# Get directors who have contributed more than 5 movies
contributing_directors = director_counts[director_counts >5].index.tolist()

# Filter india_df to include only these contributing directors
filtered_india_df = india_df[india_df['director'].isin(contributing_directors)]

# Calculate average viewer rating for these directors and sort
avg_rating_by_director = filtered_india_df.groupby('director')['viewer_rating'].mean().sort_values(ascending=False)

# Get the most rated director, their average rating, and their total contributions
most_rated_director_india = avg_rating_by_director.index[0]
highest_avg_rating_india = avg_rating_by_director.iloc[0]
contributions_of_top_director = director_counts[most_rated_director_india]

print(f"The most rated director in India is '{most_rated_director_india}' with an average rating of {highest_avg_rating_india:.2f} and about {contributions_of_top_director} contributions.")
```
